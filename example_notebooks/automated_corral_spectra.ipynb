```python
import time
from AaltoAtoms import CircCorral, get_latman_dest, get_atom_that_moved_the_most
from AMRL import Createc_Controller
import numpy as np
import matplotlib.pyplot as plt
```

```python
CC = Createc_Controller()

# get atom positions at beginning
Cim = np.array(CC.stm.scandata(1,4))
zconst = float(CC.stm.getparam('ZPiezoConst'))
nmx = nmy = CC.get_len_nm()
C = CircCorral(Cim, zconst, nmx, nmy)
C.subtract_plane()
C.get_region_centroids(percentile=98, edge_cutoff=0.1, show=True)
central_atom_loc = C.get_central_atom(C.centroids)

old_centroids = C.centroids
original_circle = C.r, C.c
```

```python
scan_bias_mv = 1000

plt.imshow(C.im)
plt.scatter(*C.c.T)

# take line spectra in one direction 
x1, y1 = C.c - [C.nm_to_pix(2*np.sqrt(2)), C.nm_to_pix(2*np.sqrt(2))] 
x2, y2 = C.c + [C.nm_to_pix(2*np.sqrt(2)), C.nm_to_pix(2*np.sqrt(2))] 
plt.plot([x1, x2], [y1, y2])
CC.stm.btn_vertspec_line(x1, y1, x2, y2)

# scan corral and save scan
CC.ramp_bias_mV(scan_bias_mv)
CC.stm.scanstart()
CC.stm.scanwaitfinished()
time.sleep(1)
CC.stm.quicksave()

# get atom positions
Cim = np.array(CC.stm.scandata(1,4))
zconst = float(CC.stm.getparam('ZPiezoConst'))
nmx = nmy = CC.get_len_nm()
C = CircCorral(Cim, zconst, nmx, nmy)
C.subtract_plane()
C.get_region_centroids(percentile=98, edge_cutoff=0.1, show=True)
central_atom_loc = C.get_central_atom(C.centroids)
old_centroids = C.centroids
original_circle = C.r, C.c

# take line spectra in other direction
x1, y1 = C.c - [C.nm_to_pix(2*np.sqrt(2)), -C.nm_to_pix(2*np.sqrt(2))] 
x2, y2 = C.c + [C.nm_to_pix(2*np.sqrt(2)), -C.nm_to_pix(2*np.sqrt(2))] 
plt.plot([x1, x2], [y1, y2])
CC.stm.btn_vertspec_line(x1, y1, x2, y2)

# scan corral and save scan
CC.ramp_bias_mV(scan_bias_mv)
CC.stm.scanstart()
CC.stm.scanwaitfinished()
time.sleep(1)
CC.stm.quicksave()

# get atom positions
Cim = np.array(CC.stm.scandata(1,4))
zconst = float(CC.stm.getparam('ZPiezoConst'))
nmx = nmy = CC.get_len_nm()
C = CircCorral(Cim, zconst, nmx, nmy)
C.subtract_plane()
C.get_region_centroids(percentile=98, edge_cutoff=0.1, show=True)
central_atom_loc = C.get_central_atom(C.centroids)
old_centroids = C.centroids
original_circle = C.r, C.c

# change bias to high bias (1.5V)
CC.ramp_bias_mV(1500)

# change lock-in amplitude to 30 mV (3000 mV/100 per Viliam's rule)
CC.stm.setp("BASICPARAM.SET", ("LockinAmpl", 30))

# take line spectra in one direction 
x1, y1 = C.c - [C.nm_to_pix(2*np.sqrt(2)), C.nm_to_pix(2*np.sqrt(2))] 
x2, y2 = C.c + [C.nm_to_pix(2*np.sqrt(2)), C.nm_to_pix(2*np.sqrt(2))] 
plt.plot([x1, x2], [y1, y2])
CC.stm.btn_vertspec_line(x1, y1, x2, y2)

# scan corral and save scan
CC.ramp_bias_mV(scan_bias_mv)
CC.stm.scanstart()
CC.stm.scanwaitfinished()
time.sleep(1)
CC.stm.quicksave()

# get atom positions
Cim = np.array(CC.stm.scandata(1,4))
zconst = float(CC.stm.getparam('ZPiezoConst'))
nmx = nmy = CC.get_len_nm()
C = CircCorral(Cim, zconst, nmx, nmy)
C.subtract_plane()
C.get_region_centroids(percentile=98, edge_cutoff=0.1, show=True)
central_atom_loc = C.get_central_atom(C.centroids)
old_centroids = C.centroids
original_circle = C.r, C.c

# take line spectra in other direction
x1, y1 = C.c - [C.nm_to_pix(2*np.sqrt(2)), -C.nm_to_pix(2*np.sqrt(2))] 
x2, y2 = C.c + [C.nm_to_pix(2*np.sqrt(2)), -C.nm_to_pix(2*np.sqrt(2))] 
plt.plot([x1, x2], [y1, y2])
CC.stm.btn_vertspec_line(x1, y1, x2, y2)
```
